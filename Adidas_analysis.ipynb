```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 

```

```python
df=pd.read_excel(r"E:\Project\Adidas US Sales Datasets.xlsx")
```

```python
df.head(5)
```

```python
df = df.copy()

```

```python
# Assign row 4 as header
df.columns = df.iloc[3]
df
```

```python

df = df.iloc[5:]

df
```

```python


df.reset_index(drop=True, inplace=True)
```

```python
df
```

```python
df.dropna(how='all', inplace=True)

df.dropna(axis=1, how='all', inplace=True)
```

```python
df
```

```python
df.columns = (
    df.columns
      .astype(str)
      .str.strip()
      .str.replace(" ", "_")
      .str.lower()
)

print(df.columns)
```

```python
print(df.dtypes)
```

```python
df['invoice_date'] = pd.to_datetime(
    df['invoice_date'],
    errors='coerce'
)

numeric_cols = [
    'price_per_unit',
    'units_sold',
    'total_sales',
    'operating_profit',
    'operating_margin'
]

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df.head(3)
```

```python
print(df.isnull().sum())

df.dropna(inplace=True)
```

```python
print(df.duplicated().sum())

df.drop_duplicates(inplace=True)
```

```python
df.to_excel(
    "cleaned_Adidas_Sales.xlsx",
    index=False
)
```
